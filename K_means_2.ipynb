```python
import pandas as ps
import numpy as np
import seaborn as sb
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#Modelo Planea
datos = ps.read_csv('/content/planea.csv', delimiter = ';')
```

```python
datos
```

```python
datos.head(1)
```

```python
datos2 = datos.drop(['NOMBRE', 'TURNO'], axis=1)
```

```python
scaler = datos
estandariza = StandardScaler()
datos2 = estandariza.fit_transform(datos2)
```

```python
#3 clusters es decir 5 grupos
kmeans = KMeans(n_clusters=5, n_init=10)
kmeans.fit(datos2)
```

```python
kmeans.labels_
```

```python
datos3 = datos.assign(Grupo = kmeans.labels_)
```

```python
datos3.sort_values(by='Grupo')
```
