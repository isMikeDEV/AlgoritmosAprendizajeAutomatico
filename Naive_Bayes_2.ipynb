```python
datos = ps.read_csv('/content/spam_aleatorio.csv', delimiter = ',')
```

```python
#Variable SPAM corresponde a si es determina o no como SPAM
datos 
```

```python
datos2 = datos[['make', 'address', 'all', 'threeD', 'our', 'over', 'remove', 'internet', 'order', 'mail', 'receive', 'will', 'people', 'report', 'addresses', 'free', 'business', 'email', 'you', 'credit', 'your', 'font', 'n000', 'money', 'hp', 'hpl', 'george', 'n650', 'lab', 'labs', 'telnet', 'n857', 'data', 'n415', 'n85', 'technology', 'n1999', 'parts', 'pm', 'direct', 'cs', 'meeting', 'original', 'project', 're', 'edu', 'table', 'conference', 'semicolon', 'leftparen', 'leftbrack', 'bang', 'dollar', 'hash', 'crl.av', 'crl.long', 'crl.tot', 'spam']]
```

```python
#Siendo Spam la variable de la tabla a predecir
y_train = X_train['spam']
y_test = X_test['spam']
```
