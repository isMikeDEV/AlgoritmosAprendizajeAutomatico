```python
import pandas as ps
import numpy as np
import seaborn as sb
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#Modelo Universidades
datos = ps.read_csv('/content/Universidades.csv', delimiter = ',')
```

```python
datos
```

```python
datos.head(1)
```

```python
datos2 = datos.drop(['Universidad'], axis=1)
```

```python
scaler = datos
estandariza = StandardScaler()
datos2 = estandariza.fit_transform(datos2)
```

```python
#3 clusters es decir 3 grupos
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(datos2)
```

```python
#Grupo  correspondientes a las universidades
kmeans.labels_
```

```python
#Asignamos a datos 3, la columna correspondiente de grupo
datos3 = datos.assign(Grupo = kmeans.labels_)
```

```python
#Mostramos de manera ordenada correspondiente al grupo
datos3.sort_values(by='Grupo')
```
